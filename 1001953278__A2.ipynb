```python
# Develop complete solution to record 10 students name, ID and mark for each subject 

import pandas as pd
Data = pd.read_csv('Student Record.csv')
print(Data)

class Student:
    def __init__(self, id:int, name:str, coursemarks:dict):
        self.ID = id
        self.Name = name
        self.CourseMarks = coursemarks
        self.GPA = 0.0

    def CalculateGPA(self):
        GradePoint = 0.0
        CreditPoint = 13
        for key,i in self.CourseMarks.items():
            if 80 <= i[0] <= 100:
                GradePoint += 4 * i[1]
                self.CourseMarks[key].append('A')
            elif 75 <= i[0] <= 79:
                GradePoint += 3.67 * i[1]
                self.CourseMarks[key].append('A-')
            elif 70 <= i[0] <= 74:
                GradePoint += 3.33 * i[1]
                self.CourseMarks[key].append('B+')
            elif 65 <= i[0] <= 69:
                GradePoint += 3 * i[1]
                self.CourseMarks[key].append('B')
            elif 60 <= i[0] <= 64:
                GradePoint += 2.67 * i[1]
                self.CourseMarks[key].append('B-')
            elif 55 <= i[0] <= 59:
                GradePoint += 2.33 * i[1]
                self.CourseMarks[key].append('C+')
            elif 50 <= i[0] <= 54:
                GradePoint += 2 * i[1]
                self.CourseMarks[key].append('C')
            elif 40 <= i[0] <= 49:
                GradePoint += 1.67 * i[1]
                self.CourseMarks[key].append('C-')
            elif 30 <= i[0] <= 39:
                GradePoint += 1.33 * i[1]
                self.CourseMarks[key].append('D')
            elif 0 <= i[0] <= 29:
                GradePoint += 0 * i[1]
                self.CourseMarks[key].append('F')

            self.GPA = GradePoint / CreditPoint

    def getID(self):
        return self.ID

    def getName(self):
        return self.Name

    def getCouseMarks(self):
        return self.CouseMarks

    def getGPA(self):
        return self.GPA

    def __str__(self):
        self.CalculateGPA()
        marks = ""
        for key, value in self.CouseMarks.items():
            marks += "{:<20}: ".format(key) + str(value[0]) + " " + str(value[2]) + "\n\t"

        return f"\nID: {self.ID}\nName: {self.Name}\nCourse Marks:\n\t{marks}" + "GPA: {:.2f}".format(self.GPA)
        

if __name__ == '__main__':
    Student_Recode = []
    while True:
        print("\n1. Student Data Entry \n2. Display All Data \n3. Display Student by ID \n4. Exit\n")
        op = int(input("Enter your choice: "))
        if op == 1:
            name = input("Enter Student name: ")
            ID = int(input("Enter ID: "))
            print("Enter the Course Marks:")
            while True:
                MT = int(input("Mass Transfer: "))
                if MT > 100 or MT < 0:
                    print("Enter marks between 0-100")
                    MT = int(input("Mass Transfer: "))

                HT = int(input("Heat Transfer: "))
                if HT > 100 or HT < 0:
                    print("Enter marks between 0-100")
                    HT = int(input("Heat Transfer: "))

                SC = int(input("Surface Chemistry: "))
                if SC > 100 or SC < 0:
                    print("Enter marks between 0-100")
                    SC = int(input("Surface Chemistry: "))

                F = int(input("Fuel: "))
                if F > 100 or F < 0:
                    print("Enter marks between 0-100")
                    F = int(input("Fuel: "))

                PC = int(input("Petro Chemical: "))
                if PC > 100 or PC < 0:
                    print("Enter marks between 0-100")
                    PC = int(input("Petro Chemical: "))

                student = Student(ID, name, {"Mass Transfer":[MT, 3],"Heat Transfer":[HT,3],"Surface Chemistry":[SC,2], "Fuel":[F,3], "Petro Chemical":[PC, 2]})
                Student_Recode.append(student)
                break

            print("***** Successfully added student! *****")

        elif op == 2:
            if len(Student_Recode) == 0:
                print("***** No Student Record Found! *****")
            for s in Student_Recode:
                s.CalculateGPA()
                print(s)
        elif op == 3:
            id = int(input("Enter student ID: "))
            for s in Student_Recode:
                if s.getID() == id:
                    print(s)
                    break
            else:
                print("***** Student not found! *****")

        elif op == 4:
            print("\n***** thanks! *****")
            exit(0)

        


```
