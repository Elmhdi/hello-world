```python
# Question 1 : Solve the following set of equations.

# numpy linear algebra

import numpy as np
a = np.array([[3, -2, 1] , [2, 6, -4] ,[-1, -2, 5]])

b = np.array ([-10, 44, -26])

x = np.linalg.solve(a,b)
x1 , x2 , x3 = x

print('x1 = %.1f'%x1)
print('x2 = %.1f'%x2)
print('x3 = %.1f'%x3)
```

```python
# Question 2
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.optimize as so

Data = pd.read_csv('dataT2.csv')
X = Data.Temp
Y = Data.Press

# plot all the poitnts from data (Expermintal data)
plt.plot(X,Y,'.b')
plt.xlabel('Temperature, T [K]')
plt.ylabel('Pressure, p [pa]')
plt.title('Pressure vs Temperature')
plt.grid()
plt.show()

def P(T,R):
 V = 10
 n = 1000/4
 return (n/V)*R*T

# caculate the coefficients
a_sol, c = so.curve_fit(P,X,Y)
r = a_sol
print(r)

# Prediction data 
Y_head = P(X,r)
plt.plot(X,Y,'.r',label = 'Y_data')
plt.plot(X,Y_head,'.b',label = 'Y_head')
plt.xlabel('Temperature, T [K]')
plt.ylabel('Pressure, P [Pa]')
plt.title('Pressure vs Temperature')
plt.grid()
plt.legend()
plt.show()

#Regeration
Y_head = P(X,r)
plt.plot(Y,Y,'r',label = 'Y_data')
plt.plot(Y,Y_head,'.g',label = 'Y_head')
plt.xlabel('Pressure')
plt.ylabel('Pressure Head')
plt.title('Pressure Head vs Preassure (Regression)')
plt.grid()
plt.legend()
plt.show()

# Calculate the error of pressure between prediction and expermental data
error = Y - Y_head
plt.bar(X,error,width = 2)
plt.xlabel('Temperature, T [K]')
plt.ylabel('Error')
plt.title('Error Bar')
plt.show()


```

```python
data = [Data['Temp'], Data['Press'],  error]
headers = ['Temperature' , 'Pressure True' , 'Pressure Predict' , 'Error']

errortable = pd.concat(data, axis=1, keys=headers)
print(errortable)
```
